```python
# Imports
import numpy as np
import math

import sys
sys.path.append("..")
from src.manifolds.torus import Torus
```

```python
# Unit tests for project_to_manifold method

# Point already on the torus
# Create new torus object
R = 3
r = 1
torus = Torus(R, r)
# Set values for u, v angles on the torus
u = 0
v = 0
# Parametrize point on torus and project back to torus
point = torus.parametrize(u, v)
updated_point = torus.project_to_manifold(point)
# Verify that both points are equivalent within small tolerance
assert np.allclose(point, updated_point, atol=1e-5)

# Point is outside of the tube
torus_point = torus.parametrize(u, v)
torus_point_normal = torus.normal_vector(u, v)

point_outside = torus_point + 0.1 * torus_point_normal
new_torus_point = torus.project_to_manifold(point_outside)

assert np.allclose(torus_point, new_torus_point, atol=1e-5)

# Point is inside the tube
point_inside = torus_point - 0.1 * torus_point_normal
new_torus_point = torus.project_to_manifold(point_inside)

assert np.allclose(torus_point, new_torus_point, atol=1e-5)

# Returned point satisfies torus equation
projected_point = torus.project_to_manifold(point_outside)

x = projected_point[0]
y = projected_point[1]
z = projected_point[2]

lhs = (math.sqrt(x**2 + y**2) - R)**2 + z**2
r_squared = r**2

assert np.allclose(lhs, r_squared, atol=1e-5)
```
